Label Encoding the data types Objects

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
for i in X_train:
       if (X_train[i].dtype!='int64'):
            X_train[i]= label_encoder.fit_transform(X_train[i])
            print(i)
            print(X_train[i].unique())
```

```python
for i in df_testdata:
       if (df_testdata[i].dtype!='int64'):
            df_testdata[i]= label_encoder.fit_transform(df_testdata[i])
            print(i)
            print(df_testdata[i].unique())
```

Performing dimentionality reduction using standard scalar

```python
std_scalar = preprocessing.StandardScaler()
X_train=std_scalar.fit_transform(X_train)
df_testdata=std_scalar.fit_transform(df_testdata)
```

Applying PCA

```python
from sklearn.decomposition import PCA
pca = PCA()
X_train_pca = pca.fit_transform(X_train)
df_testdata_pca = pca.fit_transform(df_testdata)
pca.explained_variance_ratio_
```

**Setlling down to n-comp = 200 for enough data and less varience**
> Approached by depreciating 50 from 350 (No changes in values in less than 200)

```python
pca=PCA(n_components=200)
X_train_pca=pca.fit_transform(X_train)
df_testdata_pca=pca.fit_transform(df_testdata)
pca.explained_variance_ratio_
```

Number of decision trees restriced to 200 to avoid overfitting

```python
import xgboost as xgb
xgb_model = xgb.XGBRegressor(objective ='reg:linear', n_estimators = 200)
xgb_model.fit(X_train_pca,Y_train)
```

```python
y = xgb_model.predict(df_testdata_pca)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
y_train_pred_xgb2=xgb_model.predict((X_train_pca))
np.sqrt(mean_squared_error(Y_train ,y_train_pred_xgb2))
```

```python
r2_score(Y_train ,y_train_pred_xgb2)
```
